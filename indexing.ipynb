```python
import geopandas as gpd
import matplotlib.pyplot as plt
import contextily as cxt
```

three types of indexing available:
- Selection by label (.loc)
- selection by poision (.iloc) &
- using [] as a callable indexer

```python
gdf = gpd.read_file("data/ancient_cities.geojson")
gdf.head(2)
```

using loc and iloc

```python
gdf.index
```

```python
# .loc
gdf.loc[[34, 45, 130]]
```

```python
# .loc with label to specify columns
gdf.loc[[34,78], ['Country', 'Latitude', 'Longitude']]
 
```

```python
# multiple rows with specific columns
gdf.loc[4:55, ['AD_500', 'Country']]
```

```python
# filetering with a boolean array
gdf[gdf.loc[:, "AD_1000"] > 0]
```

```python
# filtering with a boolean array plus chosing columns
gdf[gdf.loc[:, "AD_1000"] > 1000] [['Country', 'AD_1000']]
```

```python
# filtering with a boolean array plus chosing columns
gdf[gdf.loc[:, "AD_1000"] > 1000] [['Country']].reset_index()
```
